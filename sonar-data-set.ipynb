```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/kaggle/input/sonar-dataset/sonar.csv',header=None,skiprows=1)
df
```

```python
x=df.drop(60,axis=1)
y=df[60]
```

```python
y
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
y=encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

# Building Deep learning model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
model=Sequential()
model.add(Dense(units=80,input_dim=60,activation='relu')),
model.add(Dense(units=30,activation='relu')),
model.add(Dense(units=15,activation='relu')),
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
h=model.fit(x_train,y_train,epochs=10,batch_size=8,validation_split=0.3)
```

```python
loss , acc =model.evaluate(x_test,y_test)
print('Test Accuracy:',acc*100)
```

```python
loss,acc=model.evaluate(x_train,y_train)
print('Train Accuracy:',acc*100)
```

# Model is Overfitting

```python
h.history
```

```python
import matplotlib.pyplot as plt
```

```python
# Set ggplot style
plt.style.use('ggplot')

# Generate an array of epoch numbers
epochs = np.arange(0, len(h.history['loss']))

# Create a figure
plt.figure(figsize=(10, 6))

# Plot training loss with zig-zag effect
plt.plot(epochs, h.history['loss'], label='Training Loss', linestyle='-', marker='o')

# Plot validation loss with zig-zag effect
plt.plot(epochs, h.history['val_loss'], label='Validation Loss', linestyle='-', marker='x')

# Adding title and labels
plt.title('Training and Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Display the plot
plt.show()
```

# Adding L2 Regularization

```python
model=Sequential()
model.add(Dense(units=80,input_dim=60,activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.1))),
model.add(Dense(units=30,activation='relu')),
model.add(Dense(units=15,activation='relu')),
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
h=model.fit(x_train,y_train,epochs=10,batch_size=8,validation_split=0.3)
```

```python
loss , acc =model.evaluate(x_test,y_test)
print('Test Accuracy:',acc*100)
```

```python
loss,acc=model.evaluate(x_train,y_train)
print('Train Accuracy:',acc*100)
```

```python
from keras.layers import Dense, Dropout
model=Sequential()
model.add(Dense(units=80,input_dim=60,activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.1))),

model.add(Dense(units=30,activation='relu')),

model.add(Dense(units=15,activation='relu')),
model.add(Dropout(0.1)),
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
h=model.fit(x_train,y_train,epochs=10,batch_size=8,validation_split=0.3)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping

model=Sequential()
model.add(Dense(units=80,input_dim=60,activation='relu')),
model.add(Dense(units=30,activation='relu')),
model.add(Dense(units=15,activation='relu')),
model.add(Dense(units=1,activation='sigmoid'))


# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Define callbacks
checkpoint = ModelCheckpoint("best_model.keras", monitor='val_loss', mode='min', save_best_only=True, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
callbacks=[checkpoint,early_stopping]
model.fit(x_train,y_train,epochs=100,batch_size=17,callbacks=callbacks,validation_split=0.3)
```

```python
loss , acc =model.evaluate(x_test,y_test)
print('Test Accuracy:',acc*100)
```
